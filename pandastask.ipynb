```python
import pandas as pd
```

```python
df  = pd.read_csv('transactions.csv')
df
```

```python
#What is the total sales amount across all transactions?
df['t_amt'].sum()
```

```python
#Which month had the highest total transaction amount?
df["t_date"] = pd.to_datetime(df["t_date"])
df["Month"] = df["t_date"].dt.month
df.groupby("Month")["t_amt"].sum().sort_values(ascending=False).head(1)

```

```python
#What is the average transaction amount per customer?
df.groupby('cust_id')['t_amt'].mean()
```

```python
#What is the trend of total sales over months?
df.groupby("Month")["t_amt"].sum()
```

```python
#What is the highest single transaction amount recorded?
df['t_amt'].max()
```

```python
#Which service category contributes the most to total revenue?
df.groupby('services')['t_amt'].sum().sort_values(ascending=False).head(1)
```

```python
#Which product generated the highest revenue?
df.groupby('products_used')['t_amt'].sum().sort_values(ascending=False).head(1)
```

```python
#How does the average transaction amount vary between services?
df.groupby("services")["t_amt"].mean()
```

```python
#How many unique customers made purchases?
df['cust_id'].nunique()
```

```python
#Which customers have spent the most overall?
df.groupby("cust_id")["t_amt"].sum().sort_values(ascending=False).head(1)
```

```python
#What is the average number of transactions per customer?
df.groupby("cust_id").size().mean()
```

```python
#Are there customers who made multiple purchases in different categories?
df.groupby("cust_id")["services"].nunique()
```

```python
#What percentage of customers are repeat buyers?
repeat = df.groupby("cust_id").size()
(repeat[repeat > 1].count() / repeat.count()) * 100
```

```python
#Which product category has the highest total sales?
df.groupby('products_used')['t_amt'].sum().sort_values(ascending=False).head(1)
```

```python
#Which services are most popular by transaction count?
df["services"].value_counts()
```

```python
#For each service, which product type is purchased most frequently?
df.groupby(["services","products_used"]).size().reset_index(name="count") \
.sort_values(["services","count"], ascending=[True, False]) \
.drop_duplicates("services")
```

```python
#What is the average transaction amount per product type?
df.groupby("products_used")["t_amt"].mean()
```

```python
#Are there specific services where customers spend significantly more?
df.groupby("services")["t_amt"].mean().sort_values(ascending=False)
```

```python
#Which state has the highest total sales?
df.groupby("state")["t_amt"].sum().sort_values(ascending=False).head(1)
```

```python
#Which city records the highest number of transactions?
df["city"].value_counts().head(1)
```

```python
#What is the average spending per transaction in each state?
df.groupby("state")["t_amt"].mean()
```

```python
#Are certain services more popular in specific states?
df.groupby(["state","services"]).size()
```

```python
#Which states buy the most Outdoor Recreation products?
df[df["products_used"]=="Outdoor Recreation"].groupby("state").size()
```

```python
#Compare average spending between California and Texas customers.
df[df["state"].isin(["California","Texas"])].groupby("state")["t_amt"].mean()
```

```python
#Which quarter of the year saw the highest sales?
df["Quarter"] = df["t_date"].dt.quarter
df.groupby("Quarter")["t_amt"].sum().sort_values(ascending=False).head(1)
```

```python
#How does total sales vary month by month?
df.groupby("Month")["t_amt"].sum()
```

```python
#What is the total number of transactions per month?
df.groupby("Month").size()
```

```python
#Is there a particular season when sports equipment sales spike?
df[df["products_used"]=="Sports Equipment"] \
  .groupby(df["t_date"].dt.month).size() \
  .sort_values(ascending=False).head(1)
```

```python
#How many transactions were done using credit?
df[df["t_details"]=="credit"].shape[0]
```

```python
#What is the total revenue generated from credit transactions?
df[df["t_details"]=="credit"]["t_amt"].sum()
```

```python
#Is there a difference in average spending between credit and debit customers?
df.groupby("t_details")["t_amt"].mean()
```

```python
#Which states or cities should the company focus on for marketing high-value products?
df.groupby("state")["t_amt"].sum().sort_values(ascending=False)
```

```python
#Should more inventory be stocked for Exercise & Fitness products given their popularity?
df["products_used"].value_counts()
```

```python
#Are there product categories with high sales but low average transaction amounts?
df.groupby("products_used")["t_amt"].agg(["sum","mean"]) \
  .sort_values("sum", ascending=False)
```

```python
#Are there underperforming service categories that need promotional offers?
df.groupby("services")["t_amt"].sum().sort_values().head(1)
```
